```python
#Determinant of matrix without Numpy

print("Enter the elements of the 3x3 matrix row-wise:")

matrix = []
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

determinant = (
    a * (e*i - f*h)
    - b * (d*i - g*f)
    + c * (d*h - e*g)
)

print("\nDeterminant of the matrix:")
print(determinant)
```

```python
#Detminant of matrix using Numpy

import numpy as np

print("Enter the elements of 3x3 matrix row-wise:")

elements = []
for i in range(3):
    row = list(map(int, input(f"Enter the row {i+1}(elements seperated by space):").split()))
    elements.append(row)

matrix = np.array(elements)

deteminant = np.linalg.det(matrix)

print("\nmatrix: ")
print(matrix)

print("determinant of the given matrix is: ")
print(deteminant)
```

```python
#Inverse of the matrix without using Numpy

print("Enter the elements of the 3x3 matrix row wise:")

matrix = []
for i in range(3):
    row = list(map(int, input(f"Enter row {i+1}: ").split()))
    matrix.append(row)

a, b, c = matrix[0]
d, e, f = matrix[1]
g, h, i = matrix[2]

det = (
    a * (e*i - f*h)
    - b * (d*i - g*f)
    + c * (d*h - e*g)
)

print("\nDeterminant of the matrix:")
print(det)

if det == 0:
    print("Inverse of this matrix does not exist")
else:
    inverse = [

            [(e*i - f*h)/det, (c*h - b*i)/det, (b*f - c*e)/det],
            [(f*g - d*i)/det, (a*i - c*g)/det, (c*d - a*f)/det],
            [(d*h - e*g)/det, (b*g - a*h)/det, (a*e - b*d)/det]

    ]

print("\nInverse of the matrix is: ")
for row in inverse:
    print(row)
```

```python
#Solving system of Linear equation without using Numpy

print("Solve a system of 2 equations of the form:")
print("a1*x + b1*y = c1")
print("a2*x + b2*y = c2")

a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

detA = a1*b2 - a2*b1

if detA == 0:
    print("The system has no unique solution (determinant is zero).")
else:
    det_x = c1*b2 - c2*b1
    det_y = a1*c2 - a2*c1

    x = det_x / detA
    y = det_y / detA

    print("\nSolution of the system:")
    print(f"x = {round(x, 2)} , y = {round(y, 2)}")
```

```python
# Solve system of 2 linear equations using NumPy

import numpy as np

print("Solve the system:")
print("a1*x + b1*y = c1")
print("a2*x + b2*y = c2")

# Input values
a1 = float(input("Enter a1: "))
b1 = float(input("Enter b1: "))
c1 = float(input("Enter c1: "))

a2 = float(input("Enter a2: "))
b2 = float(input("Enter b2: "))
c2 = float(input("Enter c2: "))

# Create coefficient matrix A
A = np.array([[a1, b1],
              [a2, b2]])

# Create constant matrix B
B = np.array([c1, c2])

# Check determinant
det = np.linalg.det(A)

if det == 0:
    print("\nThe system has no unique solution (Determinant is zero).")
else:
    solution = np.linalg.solve(A, B)

    print("\nSolution of the system:")
    print("x =", solution[0])
    print("y =", solution[1])
```
