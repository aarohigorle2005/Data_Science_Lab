Program to calculate the Mean, Median, Mode, Variance and Standard Deviation of a dataset.

```python
# Given data
data = [10, 20, 20, 30, 40]
n = len(data)

# Mean
mean = sum(data)/n

# Median
data.sort()
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# Mode
mode = data[0]
max_count = 0
for x in data:
    count = 0
    for y in data:
        if x == y:
            count += 1
    if count > max_count:
        max_count = count
        mode = x

# Variance
variance = 0
for x in data:
    variance += (x - mean) ** 2
variance = variance / n

# Standard Deviation
std_dev = variance ** 0.5

# Output
print("Data:", data)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

 Generate a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# User input
mean = float(input("Enter mean: "))
std = float(input("Enter standard deviation: "))
n = int(input("Enter number of values: "))

# Generate normal distribution data
data = np.random.normal(mean, std, n)

# Plot histogram
plt.hist(data, bins=30, density=True)

# Plot normal distribution curve
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean, std))

# Labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.show()
```

 Compute mean, median, variance, and standard deviation for a dataset.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# User input
data = list(map(int, input("Enter numbers: ").split()))
data = np.array(data)

# Calculations
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# Histogram
plt.hist(data)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Histogram")
plt.show()
```

Simulate a coin toss experiment and compute probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Enter number of tosses: "))

# Simulate coin toss
toss = np.random.choice(["Head", "Tail"], n)

# Count results
heads = np.sum(toss == "Head")
tails = np.sum(toss == "Tail")

# Probabilities
print("Heads:", heads, " Probability:", heads/n)
print("Tails:", tails, " Probability:", tails/n)

# Graph
plt.bar(["Head", "Tail"], [heads, tails])
plt.show()
```
