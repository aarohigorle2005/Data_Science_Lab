```python
def factorial(n):
    """
    Calculates the factorial of a non-negative integer n.
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Input must be a non-negative integer.")
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage:
number = 5
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")

number = 0
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")

number = -3
try:
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def display_multiplication_table(number, limit=10):
    """
    Displays the multiplication table for a given number up to a specified limit.
    """
    if not isinstance(number, int):
        raise ValueError("Input 'number' must be an integer.")
    if not isinstance(limit, int) or limit < 1:
        raise ValueError("Input 'limit' must be a positive integer.")

    print(f"Multiplication table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
try:
    display_multiplication_table(num)
except ValueError as e:
    print(f"Error: {e}")

print("\n") # Add a newline for better separation

num = 12
limit_val = 5
try:
    display_multiplication_table(num, limit=limit_val)
except ValueError as e:
    print(f"Error: {e}")

print("\n") # Add a newline for better separation

# Example with invalid input
invalid_num = "abc"
try:
    display_multiplication_table(invalid_num)
except ValueError as e:
    print(f"Error: {e}")
```

```python
def is_palindrome_number(number):
    """
    Checks if a given number is a palindrome.
    A number is a palindrome if it reads the same forwards and backwards.
    """
    if not isinstance(number, (int, float)):
        raise ValueError("Input must be a number.")

    # Convert the number to a string to easily compare characters
    # Handle potential negative sign for comparison (e.g., -121 is not a palindrome)
    num_str = str(number)

    # For negative numbers, usually they are not considered palindromes
    # unless specified otherwise. We'll treat them as not palindromes
    # if the negative sign is part of the string.
    if num_str.startswith('-'):
        return False

    return num_str == num_str[::-1]

# Example usage:
num1 = 121
print(f"Is {num1} a palindrome? {is_palindrome_number(num1)}")

num2 = 12321
print(f"Is {num2} a palindrome? {is_palindrome_number(num2)}")

num3 = 12345
print(f"Is {num3} a palindrome? {is_palindrome_number(num3)}")

num4 = 7
print(f"Is {num4} a palindrome? {is_palindrome_number(num4)}")

num5 = -121 # Typically not considered a palindrome due to the sign
print(f"Is {num5} a palindrome? {is_palindrome_number(num5)}")

num6 = 10
print(f"Is {num6} a palindrome? {is_palindrome_number(num6)}")

# Example with invalid input
invalid_input = "hello"
try:
    print(f"Is '{invalid_input}' a palindrome? {is_palindrome_number(invalid_input)}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def check_odd_or_even(number):
    """
    Checks if a given number is odd or even.
    """
    if not isinstance(number, int):
        raise ValueError("Input must be an integer.")

    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Example usage:
num1 = 7
print(f"The number {num1} is {check_odd_or_even(num1)}.")

num2 = 10
print(f"The number {num2} is {check_odd_or_even(num2)}.")

num3 = -5
print(f"The number {num3} is {check_odd_or_even(num3)}.")

# Example with invalid input
invalid_num = 3.14
try:
    print(f"The number {invalid_num} is {check_odd_or_even(invalid_num)}.")
except ValueError as e:
    print(f"Error: {e}")
```

## Expert Systems: Concept and Components

An **expert system** is a computer program that emulates the decision-making ability of a human expert. It is designed to solve complex problems and provide decision-making capabilities in a specific domain, typically by reasoning through bodies of knowledge, represented mainly as if-then rules rather than through conventional procedural code.

### Main Components of an Expert System:

1.  **Knowledge Base:** This is the core component that stores all the domain-specific knowledge. It typically includes facts (information about the domain) and rules (IF-THEN statements that describe relationships and actions within the domain). The quality and completeness of the knowledge base are crucial for the expert system's performance.
2.  **Inference Engine:** This component acts as the "brain" of the expert system. It processes the information in the knowledge base and the facts provided by the user to draw conclusions or make recommendations. It uses various reasoning strategies, such as forward chaining (data-driven) or backward chaining (goal-driven), to apply the rules.
3.  **Working Memory (or Blackboard):** This is a dynamic database where facts about the current problem are stored during a consultation. It holds information provided by the user, intermediate conclusions drawn by the inference engine, and the final solution.
4.  **User Interface:** This component facilitates interaction between the user and the expert system. It allows users to input information, ask questions, and receive explanations or solutions from the system.

### Purpose of Expert Systems:

Expert systems aim to:
*   **Emulate human expertise:** Provide advice, diagnoses, or solutions comparable to that of a human expert.
*   **Preserve knowledge:** Capture and store valuable domain knowledge from experts, preventing its loss.
*   **Enhance decision-making:** Assist non-experts in making informed decisions in complex situations.
*   **Improve consistency:** Ensure consistent application of rules and knowledge, reducing human error.
*   **Reduce costs:** Potentially reduce the need for constant human expert consultation.

## Chosen Domain: Simple Coffee Machine Troubleshooting

For this expert system, we will use a simple domain focused on **troubleshooting common issues with a coffee machine**. This domain is straightforward enough to demonstrate the principles of IF-THEN rules without requiring extensive background knowledge.

### IF-THEN Rules for Coffee Machine Troubleshooting:

Here is a set of rules that an expert system could use to diagnose and suggest solutions for a non-functional coffee machine:

1.  **Rule 1: Power Issue**
    *   **IF** the coffee machine does not turn on
    *   **AND** the power cord is not plugged into the outlet
    *   **THEN** "Plug in the power cord."

2.  **Rule 2: No Power to Outlet**
    *   **IF** the coffee machine does not turn on
    *   **AND** the power cord is plugged in
    *   **AND** there is no power to the outlet (e.g., other devices don't work in the same outlet)
    *   **THEN** "Check the circuit breaker or try a different power outlet."

3.  **Rule 3: Water Tank Empty**
    *   **IF** the coffee machine turns on but does not brew coffee
    *   **AND** the water tank is empty
    *   **THEN** "Fill the water tank with fresh water."

4.  **Rule 4: Clogged Water Line**
    *   **IF** the coffee machine turns on but does not brew coffee
    *   **AND** the water tank is full
    *   **AND** there is no water flowing through the machine
    *   **THEN** "Descale the coffee machine to clear potential clogs."

5.  **Rule 5: Coffee Ground Issue (Too Fine/Too Coarse)**
    *   **IF** coffee tastes weak or watery
    *   **THEN** "Try using finer ground coffee or adjust the grind setting on your grinder."

6.  **Rule 6: Coffee Ground Issue (Too Much/Too Little)**
    *   **IF** coffee tastes too strong or bitter
    *   **THEN** "Try using less coffee grounds per serving."

7.  **Rule 7: Machine Needs Cleaning**
    *   **IF** coffee tastes burnt or bitter
    *   **AND** the machine has not been cleaned recently
    *   **THEN** "Clean the brewing unit and drip tray thoroughly."

## Implement Rule Engine

### Subtask:
Write Python code to create a basic rule engine. This engine should be able to store the defined rules and, given a set of facts, infer conclusions based on these rules. It should include functions to add rules and to process inputs for decision-making.


**Reasoning**:
The subtask requires defining a `Rule` class and a `RuleEngine` class with methods to add rules and infer conclusions based on facts. I will implement these classes in a single code block as per the instructions.



```python
class Rule:
    def __init__(self, conditions, conclusion):
        if not isinstance(conditions, list) or not all(isinstance(c, str) for c in conditions):
            raise ValueError("Conditions must be a list of strings.")
        if not isinstance(conclusion, str):
            raise ValueError("Conclusion must be a string.")
        self.conditions = conditions
        self.conclusion = conclusion

    def __repr__(self):
        return f"IF {' AND '.join(self.conditions)} THEN {self.conclusion}"

class RuleEngine:
    def __init__(self):
        self.rules = []

    def add_rule(self, conditions, conclusion):
        rule = Rule(conditions, conclusion)
        self.rules.append(rule)
        print(f"Rule added: {rule}")

    def infer(self, facts):
        if not isinstance(facts, list) or not all(isinstance(f, str) for f in facts):
            raise ValueError("Facts must be a list of strings.")

        conclusions = []
        for rule in self.rules:
            # Check if all conditions of the rule are present in the facts
            all_conditions_met = True
            for condition in rule.conditions:
                if condition not in facts:
                    all_conditions_met = False
                    break

            if all_conditions_met:
                conclusions.append(rule.conclusion)
        return conclusions

print("Rule and RuleEngine classes defined.")
```

## Test Expert System

### Subtask:
Test the developed expert system with various scenarios and input facts to verify that it correctly applies the rules and provides accurate decisions or diagnoses for the chosen domain.


**Reasoning**:
The user wants to test the `RuleEngine` by instantiating it, adding the predefined rules, and then running several test scenarios with different facts. This will verify the inference capabilities of the expert system.



```python
coffee_expert = RuleEngine()

# Add the rules from the previous subtask
coffee_expert.add_rule(
    conditions=["coffee machine does not turn on", "power cord is not plugged into the outlet"],
    conclusion="Plug in the power cord."
)
coffee_expert.add_rule(
    conditions=["coffee machine does not turn on", "power cord is plugged in", "no power to the outlet"],
    conclusion="Check the circuit breaker or try a different power outlet."
)
coffee_expert.add_rule(
    conditions=["coffee machine turns on but does not brew coffee", "water tank is empty"],
    conclusion="Fill the water tank with fresh water."
)
coffee_expert.add_rule(
    conditions=["coffee machine turns on but does not brew coffee", "water tank is full", "no water flowing through the machine"],
    conclusion="Descale the coffee machine to clear potential clogs."
)
coffee_expert.add_rule(
    conditions=["coffee tastes weak or watery"],
    conclusion="Try using finer ground coffee or adjust the grind setting on your grinder."
)
coffee_expert.add_rule(
    conditions=["coffee tastes too strong or bitter"],
    conclusion="Try using less coffee grounds per serving."
)
coffee_expert.add_rule(
    conditions=["coffee tastes burnt or bitter", "machine has not been cleaned recently"],
    conclusion="Clean the brewing unit and drip tray thoroughly."
)

print("\n--- Testing Scenarios ---")

# Scenario 1: Coffee machine does not turn on and power cord is not plugged in
scenario1_facts = ['coffee machine does not turn on', 'power cord is not plugged into the outlet']
print(f"\nFacts: {scenario1_facts}")
conclusions1 = coffee_expert.infer(scenario1_facts)
print(f"Conclusions: {conclusions1}")

# Scenario 2: Coffee machine turns on but does not brew, water tank is empty
scenario2_facts = ['coffee machine turns on but does not brew coffee', 'water tank is empty']
print(f"\nFacts: {scenario2_facts}")
conclusions2 = coffee_expert.infer(scenario2_facts)
print(f"Conclusions: {conclusions2}")

# Scenario 3: Coffee tastes weak or watery
scenario3_facts = ['coffee tastes weak or watery']
print(f"\nFacts: {scenario3_facts}")
conclusions3 = coffee_expert.infer(scenario3_facts)
print(f"Conclusions: {conclusions3}")

# Scenario 4: Multiple rules might apply or no rules apply based on the given facts
# (e.g., Clogged Water Line scenario)
scenario4_facts = ['coffee machine turns on but does not brew coffee', 'water tank is full', 'no water flowing through the machine']
print(f"\nFacts: {scenario4_facts}")
conclusions4 = coffee_expert.infer(scenario4_facts)
print(f"Conclusions: {conclusions4}")

# Scenario 5: Coffee tastes burnt or bitter and machine has not been cleaned recently
scenario5_facts = ['coffee tastes burnt or bitter', 'machine has not been cleaned recently']
print(f"\nFacts: {scenario5_facts}")
conclusions5 = coffee_expert.infer(scenario5_facts)
print(f"Conclusions: {conclusions5}")
```
