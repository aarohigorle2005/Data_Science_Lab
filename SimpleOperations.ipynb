Factorial of a Number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

# Taking input from user
num = int(input("Enter a number: "))

# Calling the user-defined function
result = factorial(num)

print("Factorial of", num, "is:", result)
```

Table for a given number

```python
def display_table(n):
    for i in range(1, 11):
        print(n, "x", i, "=", n * i)

# Taking input from user
num = int(input("Enter a number: "))

# Calling the user-defined function
display_table(num)
```

Palindrome

```python
def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10

    if original == reverse:
        return True
    else:
        return False

# Taking input from user
number = int(input("Enter a number: "))

# Calling the user-defined function
if is_palindrome(number):
    print("The number is a Palindrome")
else:
    print("The number is NOT a Palindrome")
```

Odd and Even Number

```python
def check_odd_even(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

# Taking input from user
number = int(input("Enter a number: "))

# Calling the user-defined function
result = check_odd_even(number)

print("The given number is:", result)
```

## Expert Systems: Concept and Components

An **expert system** is a computer program that emulates the decision-making ability of a human expert. It is designed to solve complex problems and provide decision-making capabilities in a specific domain, typically by reasoning through bodies of knowledge, represented mainly as if-then rules rather than through conventional procedural code.

### Main Components of an Expert System:

1.  **Knowledge Base:** This is the core component that stores all the domain-specific knowledge. It typically includes facts (information about the domain) and rules (IF-THEN statements that describe relationships and actions within the domain). The quality and completeness of the knowledge base are crucial for the expert system's performance.
2.  **Inference Engine:** This component acts as the "brain" of the expert system. It processes the information in the knowledge base and the facts provided by the user to draw conclusions or make recommendations. It uses various reasoning strategies, such as forward chaining (data-driven) or backward chaining (goal-driven), to apply the rules.
3.  **Working Memory (or Blackboard):** This is a dynamic database where facts about the current problem are stored during a consultation. It holds information provided by the user, intermediate conclusions drawn by the inference engine, and the final solution.
4.  **User Interface:** This component facilitates interaction between the user and the expert system. It allows users to input information, ask questions, and receive explanations or solutions from the system.

### Purpose of Expert Systems:

Expert systems aim to:
*   **Emulate human expertise:** Provide advice, diagnoses, or solutions comparable to that of a human expert.
*   **Preserve knowledge:** Capture and store valuable domain knowledge from experts, preventing its loss.
*   **Enhance decision-making:** Assist non-experts in making informed decisions in complex situations.
*   **Improve consistency:** Ensure consistent application of rules and knowledge, reducing human error.
*   **Reduce costs:** Potentially reduce the need for constant human expert consultation.

## Chosen Domain: Simple Coffee Machine Troubleshooting

For this expert system, we will use a simple domain focused on **troubleshooting common issues with a coffee machine**. This domain is straightforward enough to demonstrate the principles of IF-THEN rules without requiring extensive background knowledge.

### IF-THEN Rules for Coffee Machine Troubleshooting:

Here is a set of rules that an expert system could use to diagnose and suggest solutions for a non-functional coffee machine:

1.  **Rule 1: Power Issue**
    *   **IF** the coffee machine does not turn on
    *   **AND** the power cord is not plugged into the outlet
    *   **THEN** "Plug in the power cord."

2.  **Rule 2: No Power to Outlet**
    *   **IF** the coffee machine does not turn on
    *   **AND** the power cord is plugged in
    *   **AND** there is no power to the outlet (e.g., other devices don't work in the same outlet)
    *   **THEN** "Check the circuit breaker or try a different power outlet."

3.  **Rule 3: Water Tank Empty**
    *   **IF** the coffee machine turns on but does not brew coffee
    *   **AND** the water tank is empty
    *   **THEN** "Fill the water tank with fresh water."

4.  **Rule 4: Clogged Water Line**
    *   **IF** the coffee machine turns on but does not brew coffee
    *   **AND** the water tank is full
    *   **AND** there is no water flowing through the machine
    *   **THEN** "Descale the coffee machine to clear potential clogs."

5.  **Rule 5: Coffee Ground Issue (Too Fine/Too Coarse)**
    *   **IF** coffee tastes weak or watery
    *   **THEN** "Try using finer ground coffee or adjust the grind setting on your grinder."

6.  **Rule 6: Coffee Ground Issue (Too Much/Too Little)**
    *   **IF** coffee tastes too strong or bitter
    *   **THEN** "Try using less coffee grounds per serving."

7.  **Rule 7: Machine Needs Cleaning**
    *   **IF** coffee tastes burnt or bitter
    *   **AND** the machine has not been cleaned recently
    *   **THEN** "Clean the brewing unit and drip tray thoroughly."

Expert System

```python
def coffee_machine_expert():
    print("Coffee Machine Troubleshooting Expert System")
    print("Answer with yes or no\n")

    power = input("Is the coffee machine turning ON? (yes/no): ").lower()

    if power == "no":
        print("Expert Advice: Check the power plug or power supply.")
    else:
        water = input("Is there enough water in the machine? (yes/no): ").lower()

        if water == "no":
            print("Expert Advice: Fill the water tank.")
        else:
            coffee = input("Is coffee coming out properly? (yes/no): ").lower()

            if coffee == "no":
                print("Expert Advice: Clean the coffee outlet or check for blockages.")
            else:
                print("Expert Advice: Your coffee machine is working properly!")

# Calling the user-defined function
coffee_machine_expert()
```
